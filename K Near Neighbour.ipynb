```python
#K=1
import numpy as np
np.random.seed(0) #Using random seed constant value so we get same split of data everytime
import matplotlib.pyplot as plt
#%matplotlib inline
import mltools as ml
iris = np.genfromtxt(fname = 'C:\\Users\DELL\Desktop\dataHW1\data\iris.txt', delimiter=None)
X = iris[:, 0:2] #Using first two feature
Y = iris[:, -1]
X, Y = ml.shuffleData(X,Y) #Shuffle the data
Xtr, Xva, Ytr, Yva = ml.splitData(X,Y,0.8)  #Dividing data into train, validate 80,20 percent
print(Xtr.shape) 
print(Xva.shape)
# Creating an classifier.
knn = ml.knn.knnClassify() 
# Training the classifier.
knn.train(Xtr, Ytr, K=1)
# Making predictions
YvaHat = knn.predict(Xva)
print(YvaHat)
trainerror=knn.err(Xtr, Ytr)
print(trainerror)
validateerror =knn.err(Xva,Yva)
print(validateerror)
ml.plotClassify2D(knn, Xtr, Ytr); 
```

##  Question 3 part2

```python

import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
#%matplotlib inline
import mltools as ml
trainerror = [] #Create an empty list named as trainerror
validateerror = []  #Create an empty list validate error
iris = np.genfromtxt(fname = 'C:\\Users\DELL\Desktop\dataHW1\data\iris.txt', delimiter=None)
X = iris[:, 0:2] 
Y = iris[:, -1]
X, Y = ml.shuffleData(X,Y)
Xtr, Xva, Ytr, Yva = ml.splitData(X,Y,0.8)
print(Xtr.shape)
print(Xva.shape)
K=[1,2,5,10,50,100,200];
for k in K:
    # Creating an classifier.
    knn = ml.knn.knnClassify()
    # Training the classifier.
    knn.train(Xtr, Ytr, k)
    # Making predictions
    YvaHat = knn.predict(Xva)
    print(YvaHat)
    knn.err(Xtr, Ytr) #get training error
    knn.err(Xva,Yva)  #get validation error
    trainerror.append(knn.err(Xtr, Ytr)) #After each iteration append trainerror to list created above
    validateerror.append(knn.err(Xva,Yva)) #After each iteration append validateerror to list created above
print(trainerror)
print(validateerror)
plt.semilogx(K, trainerror, 'r-', lw=3, label='Training')
plt.semilogx(K, validateerror, 'g-', lw=3, label='Validation')
plt.legend()
```

Since we care more about validation error so I will use the K values where green plot has a dip(lower value) with fine train error and aften which the validation error begin to increase. The validation error is 0.1 at that point, and training error is 0.25 and K = 50 so I will use that value.

```python
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
#%matplotlib inline
import mltools as ml
trainerror = [] #Create an empty list named as trainerror
validateerror = []  #Create an empty list validate error
iris = np.genfromtxt(fname = 'C:\\Users\DELL\Desktop\dataHW1\data\iris.txt', delimiter=None)
X = iris[:, 0:4] #Using all features
print(X.shape)
Y = iris[:, -1]
X, Y = ml.shuffleData(X,Y)
Xtr, Xva, Ytr, Yva = ml.splitData(X,Y,0.8)
print(Xtr.shape)
print(Xva.shape)
K=[1,2,5,10,50,100,200];
for k in K:
    # Creating an classifier.
    knn = ml.knn.knnClassify()
    # Training the classifier.
    knn.train(Xtr, Ytr, k)
    # Making predictions
    YvaHat = knn.predict(Xva)
    print(YvaHat)
    knn.err(Xtr, Ytr) #get training error
    knn.err(Xva,Yva)  #get validation error
    trainerror.append(knn.err(Xtr, Ytr)) #After each iteration append trainerror to list created above
    validateerror.append(knn.err(Xva,Yva)) #After each iteration append validateerror to list created above
print(trainerror)
print(validateerror)
plt.semilogx(K, trainerror, 'r-', lw=3, label='Training')
plt.semilogx(K, validateerror, 'g-', lw=3, label='Validation')
plt.legend()
```
